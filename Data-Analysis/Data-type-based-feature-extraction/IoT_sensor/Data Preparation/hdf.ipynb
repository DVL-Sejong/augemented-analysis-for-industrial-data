```python
import pandas as pd
import numpy as np

df = pd.read_hdf('Gangnam/org_gangnam.h5')
alt_df = pd.read_csv('fully_observed_normal_speed.csv')

print(df.values.shape, alt_df.values.shape)
```

```python
alt_df = alt_df.drop(columns=['Unnamed: 0'])
alt_df.index = pd.date_range('2022-10-01', freq='5min', periods=2880)
with pd.HDFStore('fully_observed_normal_speed.h5') as store:
    store.put('/', alt_df, append=True,
             format='table', data_columns=True,
             encoding='utf-8', dropna=False)
```
