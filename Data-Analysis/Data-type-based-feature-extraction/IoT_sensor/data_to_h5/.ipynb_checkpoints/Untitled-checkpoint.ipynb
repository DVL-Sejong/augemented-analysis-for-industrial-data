```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_hdf('metr-la.h5')
df.to_csv('metr-la.csv')
df = pd.read_hdf('pems-bay.h5')
df.to_csv('pems-bay.csv')
df = pd.read_hdf('Seoul_all.h5')
df.to_csv('gangnam.csv')
```

```python
def to_hdf(file, name, idx):
    df = pd.read_csv(file + '.csv')
    alt_df = pd.read_csv(name + '/speed_' + name + '_' + idx + '.csv')
    
    print(df.values.shape, alt_df.values.shape)
    
    df.index = df['Unnamed: 0']
    df = df.drop(columns = ['Unnamed: 0'])
    df.index.name = ''
    
    alt_df = alt_df.drop(columns=['Unnamed: 0'])
    if file == 'gangnam': alt_df.index = pd.date_range('2020-10-01', periods=2880, freq='5min')
    else: alt_df.index = df.index
    alt_df.columns = df.columns
    
    print((df==0).sum().sum(), (alt_df==0).sum().sum())
    
    with pd.HDFStore(file + '_' + idx + '.h5') as store:
        store.put('/', alt_df, append=True,
                 format = 'table', data_columns = True,
                 encoding = 'utf-8', dropna=False)
```

```python
to_hdf('metr-la', 'la', '0')
```
