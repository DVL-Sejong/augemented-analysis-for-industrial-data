```python
with open('graph_sensor_ids_Seoul_all.txt') as f:
    sensor_ids = f.read().strip().replace('\n', '').replace("'", '').replace(' ', '').split(',')
distance_df = pd.read_csv('distances_Seoul_all.csv', dtype={'from':'str', 'to':'str'})
```

```python
sensor_ids
```

```python
distance_df
```

```python
normalized_k = 0.1
num_sensors = len(sensor_ids)
dist_mx = np.zeros((num_sensors, num_sensors), dtype=np.float32)
dist_mx[:] = np.inf

sensor_id_to_ind = {}
for i, sensor_id in enumerate(sensor_ids):
    sensor_id_to_ind[sensor_id] = i
```

```python
sensor_id_to_ind
```

```python
for row in distance_df.values:
    if row[0] not in sensor_id_to_ind or row[1] not in sensor_id_to_ind:
        continue
        
    dist_mx[sensor_id_to_ind[row[0]], sensor_id_to_ind[row[1]]] = row[2]
    
distances = dist_mx[~np.isinf(dist_mx)].flatten()
distances
```

```python
std = distances.std()
adj_mx = np.exp(-np.square(dist_mx/std))
adj_mx[adj_mx < normalized_k] = 0
adj_mx
```

```python
path_df = pd.DataFrame(columns=['start_lon', 'start_lat', 'end_lon', 'end_lat'])
path_df
```

```python
unique_geo = geo['sensor_id'].values
u = list(set(unique_geo))
len(u), u
```

```python
unique_dist_from = distance_df['from'].values
uu = list(set(unique_dist_from))
len(uu), u
```

```python
unique_dist_to = distance_df['to'].values
uu = list(set(unique_dist_to))
len(uu), u
```

```python
len(distance_df)
```

```python
for i in range(len(distance_df)):
    if i%100 == 0:
        print(i)
    data = distance_df.iloc[i]
    
    if data['cost'] == 0:
        continue
    
    v_from = data['from']
    v_to = data['to']
    
    if (v_from not in sensor_ids) or (v_to not in sensor_ids):
        continue
    
    g_from = None
    g_to = None
    v_t = True
    for j in range(len(geo)):
        if geo.iloc[j]['sensor_id'] == v_from:
            g_from = geo.iloc[j]
            v_t = False
        if geo.iloc[j]['sensor_id'] == v_to:
            g_to = geo.iloc[j]
    
    if v_t==True:
        continue
    
    start_lon = g_from['longitude']
    start_lat = g_from['latitude']
    end_lon = g_to['longitude']
    end_lat = g_to['latitude']
    
    info = {
        'start_lon':start_lon,
        'start_lat':start_lat,
        'end_lon':end_lon,
        'end_lat':end_lat
    }
    
    path_df = path_df.append(info, ignore_index=True)
```

```python
path_df
```

```python
for i in range(len(path_df)):
    fig.add_trace(
        go.Scattergeo(
            locationmode = 'USA-states',
            lon = [path_df['start_lon'][i], path_df['end_lon'][i]],
            lat = [path_df['start_lat'][i], path_df['end_lat'][i]],
            mode = 'lines',
            line = dict(width = 1,color = 'red'),
        )
    )
```
