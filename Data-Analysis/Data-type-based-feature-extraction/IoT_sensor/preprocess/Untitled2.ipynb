```python
keys = ['birmingham', 'bolton', 'constance', 'essen', 'groningen', 'innsbruck', 'manchester',
      'rotterdam', 'torino']
```

```python
dict(list(df.groupby('city')))['birmingham']['day'].iloc[0]
```

```python
dict(list(df.groupby('city')))['birmingham']['day'].iloc[-1]
```

```python
for key in keys:
    dfs = dict(list(df.groupby('city')))[key]
    
    print(key, dfs['speed'].isnull().sum(), len(dfs), dfs['day'].iloc[0],
         dfs['day'].iloc[-1])
```

```python
print(repr('City').ljust(12), '|', repr('Length').ljust(10), '|', repr('Null').ljust(10), '|',
      repr('Per').ljust(10), '|', repr('Start').ljust(15), '|', repr('End').ljust(15))
print('-------------------------------------------------------------------------------------')

nkeys = list(df.groupby('city').groups.keys())
for key in nkeys:
    dfs = dict(list(df.groupby('city')))[key]
    
    print(
          repr(key).ljust(12), '|', repr(str(len(dfs))).ljust(10), '|',
          repr(str(dfs['speed'].isnull().sum())).ljust(10), '|',
          repr(str(round(dfs['speed'].isnull().sum()/len(dfs)*100,3)) + '%').ljust(10), '|',
          repr(dfs['day'].iloc[0]).ljust(15), '|',
          repr(dfs['day'].iloc[-1]).ljust(15)
         )
```
