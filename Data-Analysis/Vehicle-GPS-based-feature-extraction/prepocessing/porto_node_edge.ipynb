```python
import numpy as np
import pandas as pd
```

```python
node = pd.read_csv('C:/users/skang/trajectory/network-Porto/nodes.csv')
edge = pd.read_csv('C:/users/skang/trajectory/network-Porto/edges.csv')
```

```python
i = 0
list = []
for point in node.geometry:
    temp = point[7:-1]
    la, lo = temp.split(' ')
    if float(la) < -8.6387 or float(la) > -8.5841:
        list.append(i)
        i += 1
        continue
    elif float(lo) < 41.1417 or float(lo) > 41.1722:
        list.append(i)
        i += 1
        continue
    i += 1
```

```python
list.reverse()
```

```python
for i in list:
    node = node.drop(node.index[int(i)])
node
```

```python
flag = 0
n=0
drop_list = []
for u, v in zip(edge['u'], edge['v']):
    for i in list:
        if int(u) == i:
            flag = 1
            break
        elif int(v) == i:
            flag = 1
            break
    if flag != 1:
        drop_list.append(n)
    flag = 0
    n += 1
drop_list
```
