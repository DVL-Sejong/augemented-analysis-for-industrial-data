```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv('data/train.csv')
df
```

```python
df = df.drop(columns={'TRIP_ID', 'CALL_TYPE', 'ORIGIN_CALL', 'ORIGIN_STAND', 'TAXI_ID', 'DAY_TYPE', 'MISSING_DATA'})
df
```

```python
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'], unit='s')
```

```python
df = df[df['POLYLINE'] != '[]']
```

```python
df['geometry']=''
df
```

```python
i = 0
for test in tqdm(df.POLYLINE):
    flag = 0
    temp = test[2:-2]
    list = temp.split('],[')
    wkt = 'LINESTRING('
    for t in list:
        a = t.split(',')
        if (float(a[0]) < -8.6337 or float(a[0]) > -8.5984):
            continue
        if (float(a[1]) < 41.1403 or float(a[1]) > 41.1659):
            continue
        wkt += a[0] + ' '
        wkt += a[1] + ','
    wkt = wkt[:-1]
    wkt += ')'
    df['geometry'][i] = wkt
    i += 1
df.head()
```

```python
df.to_csv('DATA/prev.csv', index=False)
```
