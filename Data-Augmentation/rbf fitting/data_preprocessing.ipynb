```python
import pandas as pd
```

```python
eeg_col = ['EEG.Cz', 'EEG.Fz', 'EEG.Fp1',  'EEG.F7', 'EEG.F3', 'EEG.FC1', 'EEG.C3','EEG.FC5',
           'EEG.FT9', 'EEG.T7','EEG.CP5',  'EEG.CP1', 'EEG.P3', 'EEG.P7', 'EEG.PO9', 'EEG.O1',
           'EEG.Pz', 'EEG.Oz', 'EEG.O2', 'EEG.PO10', 'EEG.P8', 'EEG.P4', 'EEG.CP2', 'EEG.CP6',
           'EEG.T8', 'EEG.FT10', 'EEG.FC6', 'EEG.C4', 'EEG.FC2', 'EEG.F4', 'EEG.F8', 'EEG.Fp2']
```

```python
path = path = './{}/base/{}.csv'
```

```python
df = pd.read_csv(path.format('cyj', '20'), header = 1)
```

```python
CQ_col = list(df.columns[44:76])
```

```python
def CQ_preprocessing(path, task, name, CQ_list, eeg_col):
    for n in name:
        for t in task:
            df = pd.read_csv(path.format(n, t), header = 1)
            print('Start preprocessing of {}, {}'.format(n,t))
            print('dataframe length is {}'.format(len(df)))
            for cq in CQ_list:
                df = df[df[cq] != 0]
            df[eeg_col].to_csv(path.format(n, t+'_cq'), index = False)
            print(print('preprocessing dataframe length is {}'.format(len(df))))
            print('-----------------------------------')
            print()
```

```python
name = ['cyj', 'cyy', 'jh','jsp','sbs','usb','shc']
task = ['10','20','1back','3back']
```

```python
CQ_preprocessing(path, task, name, CQ_col, eeg_col)
```

```python
def CQ_preprocessing_noise(path, task, name, CQ_list, eeg_col):
    for n in name:
        if n == 'cyj':
            for t in ['20', '1back', '3back']:
                df = pd.read_csv(path.format(n, t), header = 1)
                print('Start preprocessing of {}, {}'.format(n,t))
                print('dataframe length is {}'.format(len(df)))
                for cq in CQ_list:
                    df = df[df[cq] != 0]
                df[eeg_col].to_csv(path.format(n, t+'_cq_noise'), index = False)
                print(print('preprocessing dataframe length is {}'.format(len(df))))
                print('-----------------------------------')
                print()
        else:
            for t in task:
                df = pd.read_csv(path.format(n, t), header = 1)
                print('Start preprocessing of {}, {}'.format(n,t))
                print('dataframe length is {}'.format(len(df)))
                for cq in CQ_list:
                    df = df[df[cq] != 0]
                df[eeg_col].to_csv(path.format(n, t+'_cq_noise'), index = False)
                print(print('preprocessing dataframe length is {}'.format(len(df))))
                print('-----------------------------------')
                print()
```

```python
path_noise = './{}/noise/{}.csv'
```

```python
CQ_preprocessing_noise(path_noise, task, name, CQ_col, eeg_col)
```
