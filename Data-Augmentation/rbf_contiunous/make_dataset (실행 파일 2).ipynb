```python
import pandas as pd
import numpy as np
import os
import random
```

```python
def dataframelen(task, weight='noweight'):
    folder_path = './preprocessing/' + weight + '/' + task
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

    print(task)

    len_list = []

    for csv_file in csv_files:
        df = pd.read_csv(folder_path + '/' + csv_file)
        len_list.append(len(df))
    
    print(task + ' ' + weight + ' mean df num :', np.mean(np.array(len_list)))
```

```python
dataframelen('1back')
dataframelen('3back')
dataframelen('10')
dataframelen('20')
```

```python
dataframelen('1back', 'weight')
dataframelen('3back', 'weight')
dataframelen('10', 'weight')
dataframelen('20', 'weight')
```

```python
def seqdata(task, weight='noweight'):
    if task == '1back' or task == '10':
        label = 0
    elif task == '3back':
        label = 2
    else:
        label = 1

    folder_path = './preprocessing/' + weight + '/' + task
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]


    datas = []
    labels = []
    for csv_file in csv_files:
        df = pd.read_csv(folder_path + '/' + csv_file)
        split_datas = []
        split_index = []

        for i in range(0, len(df)-120-240, 120):
            split_datas.append(df.iloc[i+120:i+120+120].values)
            split_index.append(label)
        
        split_index = np.array(split_index)
        split_datas = np.array(split_datas)

        ind_list = [i for i in range(len(split_datas))]
        random.shuffle(ind_list)

        split_datas = split_datas[ind_list, :, :]
        split_index = split_index[ind_list]

        split_datas = split_datas.transpose(0, 2, 1)
        
        print(split_datas.shape, split_index.shape, csv_file, task, weight)
        np.save('./seq_data/' + weight + '/' + task + '/' + csv_file[:-4] + '_split_datas', split_datas)
        np.save('./seq_data/' + weight + '/' + task + '/' + csv_file[:-4] + '_split_labels', split_index)
        datas.append(split_datas)
        labels.append(split_index)
    total_datas = np.concatenate(datas, axis=0)
    total_labels = np.concatenate(labels, axis=0)

    np.save('./seq_data/' + weight + '/' + task + '/' + 'total' + '_split_datas', total_datas)
    np.save('./seq_data/' + weight + '/' + task + '/' + 'total' + '_split_labels', total_labels)
```

```python
seqdata('1back')
print('-----------------------------')
seqdata('10')
print('-----------------------------')
seqdata('20')
print('-----------------------------')
seqdata('3back')
print('-----------------------------')
seqdata('1back', weight='weight')
print('-----------------------------')
seqdata('10', weight='weight')
print('-----------------------------')
seqdata('20', weight='weight')
print('-----------------------------')
seqdata('3back', weight='weight')
print('-----------------------------')
```

```python
seqdata_nonscale('1back')
print('-----------------------------')
seqdata_nonscale('10')
print('-----------------------------')
seqdata_nonscale('20')
print('-----------------------------')
seqdata_nonscale('3back')
print('-----------------------------')
seqdata_nonscale('1back', weight='weight')
print('-----------------------------')
seqdata_nonscale('10', weight='weight')
print('-----------------------------')
seqdata_nonscale('20', weight='weight')
print('-----------------------------')
seqdata_nonscale('3back', weight='weight')
print('-----------------------------')
```
