```python
def rbf_train_flow(model):
    aaa = []
    for j in range(9):
        center = model.train_center_list[j*7:(j+1) * 7]
        sigma = model.train_sigma_list[j*7:(j+1) * 7]
        weight = model.train_weight_list[j*7:(j+1) * 7]
        rbf_list = []
        for i in range(len(center)):
            rbf_output = torch.exp(-1 * (torch.pow((input_data - center[i]), 2) / \
                                            (torch.pow(sigma[i], 2))))
            pred = torch.mm(weight[i], rbf_output)
            rbf_list.append(pred)
        aaa.append(rbf_list)
    for i in range(9):
        if i ==0:
            a = torch.cat(aaa[i], axis = 0).reshape(7,5,100)
        else:
            a += torch.cat(aaa[i], axis = 0).reshape(7,5,100)

    return a
```

```python
aaa = rbf_train_flow(model_short)
bbb = rbf_train_flow(model)
```

```python
import matplotlib.pyplot as plt
num_feature = 0
for i in range(len(aaa)):
    plt.figure(figsize=(10, 4))
    plt.scatter(input_data.cpu().detach().numpy(), target_ground2[num_feature].cpu().detach().numpy(), c = 'green')
    plt.scatter(input_data.cpu().detach().numpy(), target2[num_feature].cpu().detach().numpy(), c = 'blue')
    plt.plot(input_data.cpu().detach().numpy(), aaa[i][num_feature].cpu().detach().numpy(), c = 'red')
    #plt.ylim(-5,20)
    # plt.xlim(35, 70)
    plt.show()
```
