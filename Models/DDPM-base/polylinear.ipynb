```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv('C:/Users/chanyoung/Desktop/RBFfitting/data/linear_SEM/linear_random_15.csv')
```

```python
data_df
```

```python
from scipy import interpolate
import matplotlib.pyplot as plt
```

```python
a = list()
for col in data_df.columns:
    y = data_df[col].values
    x = np.array(list(data_df.index), dtype = float)
    xnew = np.arange(0,99, 0.5)
    f_quadr = interpolate.interp1d(x, y, kind='quadratic')
    y_new_quadr = f_quadr(xnew)
    a.append([y_new_quadr])
new_df = pd.DataFrame(np.concatenate(a).T, columns = data_df.columns)
new_df
```

```python
new_df.to_csv('./data/poly_linear_200.csv', index = False)
```
